# ATM Machine Simulation

```python
Balance = 1000
PIN = 1234
attempts = 0

# PIN verification.
while attempts < 3:
    user_PIN = int(input("Enter the PIN: "))
    if user_PIN == PIN:
        print("Access Granted")
        break
    else:
        attempts = attempts + 1  
        print("Access Denied")
    
else:  
    print("Too many incorrect attempts, exiting...")
    exit()

# Menu loop.
while True:   
    menu = int(input("\nChoose an option:\n1. Check Balance\n2. Deposit\n3. Withdraw\n4. Exit\nEnter choice: "))
    if menu == 1:
        print("Your Current Balance is: ", Balance)
    elif menu == 2: 
        Deposit = int(input("Enter Amount to Deposit (INR): "))
        Balance = Balance + Deposit
        print("Your Updated Balance is: ", Balance)
    elif menu == 3:
        Withdraw = int(input("Enter Amount to Withdraw (INR): "))
        if Withdraw > Balance:
            print("Insufficient Balance!")
        else:
            Balance = Balance - Withdraw
            print("Your Updated Balance is: ", Balance)
    elif menu == 4:
        print("Exit")
        break
    else:
        print("Invalid choice, Please try again.")
```
