# Mini Project 1 - Library Management System (LMS) (Python Basics)

```python
# Welcome message. 
print("Welcome to our Library Management System (LMS)!\n") # \n adds a blank line after the welcome message is printed. 

# Default login credentials: username and password. 
user_login_credentials = {"Username": "LMS_Admin$321", "Password": "BookVerse@Admin"}

attempts = 0 # Initializes a counter to keep track of how many login attempts the user has made.

while attempts < 3: # while loop runs allowing 3 attempts to the user for entering the login credentials. 
    username = input("Enter username: ")
    password = input("Enter password: ")
    
    if username == user_login_credentials["Username"] and password == user_login_credentials["Password"]: # Compares the entered username and password with the default values in the "user_login_credentials" dictionary. "and" ensures both username and password must match.
        print("Access Granted!\nThank You!")
        break # Exits the while loop immediately, after the login credentials match — successful login stops further attempts.
    else:
        print("Access Denied!\nTry Again...\nThank You!")
        attempts = attempts + 1 # Increments the attempts counter by 1.

else: # This else is paired with the while loop. It executes only if the while loop finishes normally (i.e., the loop condition became false) — which happens when attempts reaches 3 and break was never executed.
    print("Too many incorrect attempts, exiting...!\nThank You!") # Prints a message about too many attempts. 
    exit()

# Creating a dictionary "BookVerse".
BookVerse = {
    "Fluent Python": {"Author": "Luciano Ramalho", "Availability": True, "Issued_to": None},
    "Introduction to Machine Learning with Python": {"Author": "Sarah Guido", "Availability": True, "Issued_to": None},
    "Deep Learning with Python": {"Author": "Francois Chollet", "Availability": True, "Issued_to": None},
    "Natural Language Understanding": {"Author": "James Allen", "Availability": True, "Issued_to": None},
    "Generative Deep Learning": {"Author": "David Foster", "Availability": True, "Issued_to": None}
}

# Defining all the functions. 
def view_books():
    print("Our LMS BookVerse Repository:\n") # Heading, after which a blank line (\n) follows.
    for title, details in BookVerse.items(): # Iterates over each key: value pair in BookVerse. title is the "book title" string; details is the nested dictionary for that book (includes "Author", "Availability", "Issued to" keys, and their respective values).
        status = "Available" if details["Availability"] else f"Issued to {details['Issued_to']}" # if details["Availability"] is True, status becomes "Available". Otherwise, it formats the string "Issued to ..." using details["Issued_to"].
        print(f"{title} — Author: {details['Author']} | Status: {status}") # Prints the book title, author, and the status determined above. "|" is just a text separator here.  

def add_book():
    new_entry = input("Enter the book title and the author name (separated by a comma): ")
    title, author = [x for x in new_entry.split(",")] # Splits the input string at the comma into 2 parts, and assigns to title and author.
    if title in BookVerse:
        print("The book already exists in BookVerse.\nThank You!")
    else:
        BookVerse[title] = {"Author": author, "Availability": True, "Issued_to": None} # Adds a new entry to BookVerse. 
        print(f"The book {title} has been added successfully.\nThank You!")

def issue_book():
    entry = input("Enter the book title: ")
    if entry in BookVerse:
        if BookVerse[entry]["Availability"] == True: # If the book is currently available, then issue the book as below. 
            user = input("Enter your name: ") # Ask for the name of the person issuing the book.
            BookVerse[entry]["Availability"] = False # sets the Availability to False, so the book is no longer available (i.e., it has been issued).
            BookVerse[entry]["Issued_to"] = user # Stores who has issued the book (in the "Issued_to" key).
            print(f"The book {entry} has been issued to {user}.\nThank You!")
        else:
            print(f"The book {entry} has already been issued to {BookVerse[entry]['Issued_to']}.\nThank You!") # Prints a message saying who already has the book.
    else:
        print("Book not found.\nThank You!")

def return_book():
    entry = input("Enter the book title to return: ")
    if entry in BookVerse: # Checks if the book is available in BookVerse. 
        if BookVerse[entry]["Availability"] == False: # If the book is currently not available (i.e., is issued)...
            BookVerse[entry]["Availability"] = True # Assigns True when the book is returned and hence is available.
            BookVerse[entry]["Issued_to"] = None # Issued_to field is cleared, by updating it with the value "None".
            print(f"The book {entry} has been returned.\nThank You!")
        else:
            print(f"The book {entry} was not issued.\nThank You!")
    else:
        print("Book not found.\nThank You!") 

def view_issued_books():
    print("\nIssued Books: ")
    issued = False # Indicates that no books are currently issued. 
    for title, details in BookVerse.items():
        if details["Availability"] == False: # If a book is not available (i.e., is issued)...
            print(f"{title} — Issued to {details['Issued_to']}") # Prints the title and who it is issued to (the "Issued_to" key is updated). 
            issued = True # Marks that at least one issued book exists.
    if not issued: 
        print("No books are currently issued.\nThank You!")

while True: # Starts an infinite loop for the interactive menu. The loop runs until a break occurs.
    print("\nMenu Options:")
    print("1. View Books")
    print("2. Add a Book")
    print("3. Issue a Book")
    print("4. Return a Book")
    print("5. View Issued Books")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ") # Reads the user's menu choice.
    
    if choice == "1":
        view_books() # "1" calls view_books().
    elif choice == "2":
        add_book() # "2" calls add_book().
    elif choice == "3":
        issue_book() # "3" calls issue_book().
    elif choice == "4":
        return_book() # "4" calls return_book().
    elif choice == "5":
        view_issued_books() # "5" calls view_issued_books().
    elif choice == "6":
        print("Exiting...!\nThank You!") # "6" prints exit message and breaks out of the loop (ending the program).
        break
    else:
        print("Invalid choice, please try again.\nThank You!")
```
